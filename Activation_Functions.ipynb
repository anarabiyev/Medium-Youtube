<h1>Sigmoid

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate values from -7 to 7
x_values = np.linspace(-7, 7, 200)

# Apply the sigmoid function to the x values
y_values = sigmoid(x_values)

# Plot the sigmoid curve
plt.plot(x_values, y_values, label='Sigmoid Function')
plt.title('Sigmoid Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.grid(True)
plt.show()
```

<h1>ReLU

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x)

# Generate values from -5 to 5
x_values = np.linspace(-5, 5, 200)

# Apply the ReLU function to the x values
y_values = relu(x_values)

# Plot the ReLU function
plt.plot(x_values, y_values, label='ReLU Function')
plt.title('ReLU Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```

<h1>Leaky ReLU

```python
import numpy as np
import matplotlib.pyplot as plt

def leaky_relu(x, alpha=0.05):
    return np.maximum(alpha * x, x)

# Generate values from -5 to 5
x_values = np.linspace(-5, 5, 200)

# Apply the Leaky ReLU function to the x values
y_values = leaky_relu(x_values)

# Plot the Leaky ReLU function
plt.plot(x_values, y_values, label='Leaky ReLU Function (alpha=0.05)')
plt.title('Leaky ReLU Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```

<h1>Tanh

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the tanh function
def tanh(x):
    return np.tanh(x)

# Generate values from -5 to 5
x_values = np.linspace(-5, 5, 200)

# Apply the tanh function to the x values
y_values = tanh(x_values)

# Plot the tanh function
plt.plot(x_values, y_values, label='tanh Function')
plt.title('tanh Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```

<h1>Sigmoid and its Derivative

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Define the derivative of the sigmoid function
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Generate values from -7 to 7
x_values = np.linspace(-7, 7, 200)

# Apply the derivative of the sigmoid function to the x values
y_values = sigmoid_derivative(x_values)

# Plot the derivative of the sigmoid function
plt.plot(x_values, y_values, label="Derivative")
plt.plot(x_values, sigmoid(x_values), label="Sigmoid")
plt.title("Sigmoid Function and its Derivative")
plt.xlabel("Input")
plt.ylabel("Output")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```
