```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('insurance.csv')
```

```python
df.head(10)
```

<h1>Original Data

```python
plt.hist(df['age'], bins='auto', edgecolor='black', color = 'darkorange')
plt.show()
```

<h1>Feature sclaing (X/2)

```python
plt.hist(df['age']/2, bins='auto', edgecolor='black', color = 'darkorange')
plt.show()
```

<h1>Feature sclaing (X-20)

```python
plt.hist(df['age'] - 20, bins='auto', edgecolor='black', color = 'darkorange')
plt.show()
```

<h1>Manual Normalization

```python
df['normalized_age'] = (df['age'] - df['age'].min()) / (df['age'].max() - df['age'].min())
```

```python
plt.hist(df['normalized_age'], bins='auto', edgecolor='black', color = 'darkorange')
plt.show()
```

<h1>Sklearn Normalization

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

df['normalized_age_sk'] = scaler.fit_transform(df[['age']])
```

```python
plt.hist(df['normalized_age_sk'], bins='auto', edgecolor='black', color = 'darkorange')
plt.show()
```

<h1>Sklearn Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

df['standardized_age'] = scaler.fit_transform(df[['age']])
```

```python
plt.hist(df['standardized_age'], bins='auto', edgecolor='black', color = 'darkorange')
plt.show()
```

<h1>Standardization example with 2D data

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Set a random seed for reproducibility
np.random.seed(42)

# Generate random data for X and Y
mean_X, mean_Y = 3, 5
variance_X, variance_Y = 6, 15
size = 50

X = np.random.normal(loc=mean_X, scale=np.sqrt(variance_X), size=size)
Y = np.random.normal(loc=mean_Y, scale=np.sqrt(variance_Y), size=size)

# Create a scatter plot of the original data
plt.figure(figsize=(10, 5))
plt.scatter(X, Y, label='Original Data', color = 'limegreen')
plt.title('Original Data Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.xlim([-10, 10])  # Set X-axis limits
plt.ylim([-10, 10])  # Set Y-axis limits

plt.axhline(0, color='black', linewidth=1)  # Make y-axis bold
plt.axvline(0, color='black', linewidth=1)  # Make x-axis bold

plt.show()

# Combine X and Y into a 2D array for StandardScaler
data = np.column_stack((X, Y))

# Use StandardScaler to standardize the data
scaler = StandardScaler()
standardized_data = scaler.fit_transform(data)

# Create a scatter plot of the standardized data
plt.figure(figsize=(10, 5))
plt.scatter(standardized_data[:, 0], standardized_data[:, 1], label='Standardized Data', color = 'darkorange')
plt.title('Standardized Data Scatter Plot')
plt.xlabel('Standardized X')
plt.ylabel('Standardized Y')
plt.legend()
plt.grid(True)
plt.xlim([-10, 10])  # Set X-axis limits
plt.ylim([-10, 10])  # Set Y-axis limits

plt.axhline(0, color='black', linewidth=1)  # Make y-axis bold
plt.axvline(0, color='black', linewidth=1)  # Make x-axis bold

plt.show()
```
